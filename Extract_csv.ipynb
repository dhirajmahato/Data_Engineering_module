```python
import mysql.connector

# Establish a connection to the database
cnx = mysql.connector.connect(
    host="localhost",
    user="root",
    password="Password@123",
    database="mydatabase1"
)

# Create a cursor
cursor = cnx.cursor()

# Execute the SQL query
cursor.execute("SELECT * FROM Persons")

# Fetch the results
rows = cursor.fetchall()

# Print the results
for row in rows:
    print(row)

# Close the cursor and the connection objects
cursor.close()
cnx.close()
```

```python
import mysql.connector as mysql
import csv
#Connect DB server and database
conn = mysql.connect(user='root', password='Password@123', host='localhost', database='startupindia')
cursor = conn.cursor()

# Define the SQL command to create the table if it doesn't exist
create_table_query = """
    CREATE TABLE IF NOT EXISTS Startup_info (
        Name VARCHAR(255),
        Location VARCHAR(255),
        Industry VARCHAR(255)
    )
"""

# Execute the CREATE TABLE query
cursor.execute(create_table_query)

# Open the CSV file to be put in same directory as ipyb file
with open('startup_data_150.csv', mode='r') as csv_file:
    # Read csv using reader class
    csv_reader = csv.reader(csv_file)
    # Skip header
    next(csv_reader)
    # Read csv row wise and insert into table
    for row in csv_reader:
        Startup_Name, Location, Industry = row  # Assuming these are the columns in your CSV
        sql = "INSERT INTO Startup_info (Name, Location, Industry) VALUES (%s, %s, %s)"
        values = (Startup_Name, Location, Industry)
        cursor.execute(sql, values)
        print("Record inserted")

    # Commit the changes
    conn.commit()

# Close the cursor and the connection objects
cursor.close()
conn.close()
```
